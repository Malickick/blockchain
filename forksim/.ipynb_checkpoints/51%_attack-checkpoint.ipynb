```python
import math
import matplotlib.pyplot as plt
import matplotlib
import numpy as np 
from scipy.interpolate import spline
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
# Revoie la probabilité de succès d'un attaquant ayant une portion q 
# de la puissance de calcul du réseau pour z confirmations
def attacker_success(q, z):
    p = 1.0 - q
    lmda = (z * (q / p)) # Lambda
    s = 0.0 # Somme
    
    for k in range(z+1):
        k_fact = math.factorial(k)
        numerator = math.pow(lmda, k) * math.exp(-lmda) * (1 - math.pow(q / p, z - k))
        s += numerator / k_fact
    return 1 - s

def attacker_success_35(q, z):
    p = 1.0 - q
    lmda = (z * (q / p)) # Lambda
    s = 0.0 # Somme
    
    for k in range(z+1):
        k_fact = math.factorial(k)
        numerator = math.pow(lmda, k) * math.exp(-lmda) * (1 - ( (1 - math.pow(p / q, z + 35.0 - k)) / (1 - math.pow(p / q, z + 35.0 - k + z + 1.0 - k)) ) )
        s += numerator / k_fact
    return 1 - s

def prob_attack(q, z_max):
    x = []
    y = []
    for z in range(z_max+1):
        p = attacker_success(q, z)
        #print("Z=%d\tP=%f"%(z,p))
        x.append(z)
        y.append(p)
    return x,y
```

```python
q = 0.1

print("q = %f"%q)
for z in range(11):
    print("Z=%d\tP=%f"%(z,attacker_success(q, z)))
```

```python
x_10,y_10 = prob_attack(0.1, 35)
x_20,y_20 = prob_attack(0.2, 35)
x_30,y_30 = prob_attack(0.3, 35)
x_40,y_40 = prob_attack(0.4, 35)
x_50,y_50 = prob_attack(0.5, 35)

#x_51,y_51 = prob_attack(0.51, 40)
#x_60,y_60 = prob_attack(0.6, 40)
```
